```python
#outlier detection  and removing
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
agedata=[50,51,52,53,190,43,54,65,66,67,47,45,48]
xvalues=range(len(agedata))
plt.scatter(xvalues,agedata)
plt.xlabel("Index")
plt.ylabel("Age")
plt.title("scatter plot of age data")
plt.show()
agedata=np.round(agedata,decimals=0)
agedata=agedata.tolist()
print(agedata)
sns.set_theme()
sns.displot(agedata).set(title="Distribution of Age Data")
plt.show()
mean=np.mean(agedata)
stddev=np.std(agedata)
threshold=2
outliers1=[]
for age in agedata:
  if abs(age  -  mean)>threshold*stddev:
    outliers1.append(age)
print("outliers",outliers1)
mean=np.mean(agedata)
std=np.std(agedata)
print("mean of the data set is",mean)
print("std deviation is ",std)
threshold=3
outliers2=[]
for i in agedata:
  z=(i-mean)/std
  if abs(z)>threshold:
    outliers2.append(i)
print("outliers in dataset is",outliers2)
agedata_np=np.array(agedata)
filtered_data=agedata_np[((agedata_np-mean)/std) <threshold]
print("filtered data is",filtered_data)
```

```python
#outlier using percentile method
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
agedata=[50,51,52,53,190,43,54,65,66,67,47,45,48]
lower_limit=np.percentile(a=agedata,q=5)
upper_limit=np.percentile(a=agedata,q=95)
print("lower limit",lower_limit)
print("upper limit",upper_limit)
outlier4=[]
for x in agedata:
  if(( x>upper_limit) or (x<lower_limit)):
    outlier4.append(x)
print("outliers in the dataset is",outlier4)
agedata_np=np.array(agedata)
filtered_data=agedata_np[(agedata_np>lower_limit) & (agedata_np<upper_limit)]
print("filtered data is",filtered_data)
```
